# Activation 

## sigmoid



$sigmoid(x) = \sigma(x) = \large\frac{1}{1+e^{-x}}$

a neuron with 3 inputs

$y = \sigma(z) = \sigma( \sum\limits_{i=1}^3 w_i\cdot x_i + b)$

$loss L = d(\hat y, y)$

$x_i 是我們收集的資料，訓練或推估時就是我們已知的值$

$\sigma 是我們設定模型時的運算$

$\hat y 是我們收到資料後作的標籤，訓練時也是已知值$




````
import numpy as np
````

````
def sigmoid(x):
  x = np.array(x)
  z = 1/(1+np.exp(-x))
  return z
````

## Softmax:

$z_j = \frac{\Large e^{x_{\small j}}}{\large\sum\limits_{i=1}^n e^{x_{\small i}}},\space \space j=1, 2, 3... \large n$ 

softmax 除了輸出總和為1, 也有拉大比例的趨勢

$以 1,4 輸入$ 

會得到 $0.047, 0.953$



````
def softmax(X):  # define activation: softmax      
    X = np.array(X)
    return np.exp(X) / np.sum(np.exp(X), axis=0, keepdims=True)
````

## cce function

````
def cross_entropy(p, q):
    delta = 1e-15
    H = 0
    for i in range(len(p)):
        H += -p[i]*np.log(q[i]+delta)

    H = H.sum()/p.shape[0]
    return H
````

## MSE

````
# from keras(?)
def mean_squared_error(y_true, y_pred):
  y_true = np.array(y_true)
  y_pred = np.array(y_pred)
  output_errors = np.average((y_true - y_pred) ** 2)
  return output_errors
````

````
def MSE(true, pred):
    MSE = 0
    for i in range(len(true)):
        MSE += (true[i]-pred[i])**2
    return MSE/len(true)
````

## loss: mse

x-axis is $\hat{y}^n * f(x)$

y-axis is loss value

$l(f(x^n), \hat{y}^n) = (\hat{y}^nf(x^n) - 1)^2$
- $l(f(x^n), \hat{y}^n) = (f(x^n) - 1)^2, if \space \space \hat{y}^n = +1$
- $l(f(x^n), \hat{y}^n) = (f(x^n) + 1)^2, if \space \space \hat{y}^n = -1$




````
import numpy as np
import matplotlib.pyplot as plt

f = np.linspace(-16, 18, 1701)
def loss_mse(x, y=1):
  loss = 1*(x-1)**2
  return loss
loss = loss_mse(f,1)
plt.plot(f, loss)
plt.title('mse loss, if y is +1')
plt.grid(b=True)
````

## loss: mse(sigmoid)

x-axis is $\hat{y}^n * f(x)$

y-axis is loss value

$l(f(x^n), \hat{y}^n) = (\sigma(\hat{y}^nf(x^n)) - 1)^2$
- $l(f(x^n), \hat{y}^n) = (\sigma(f(x^n)) - 1)^2, if \space \space \hat{y}^n = +1$
- $l(f(x^n), \hat{y}^n) = (\sigma(f(x^n)))^2, if \space \space \hat{y}^n = -1$

````

f = np.linspace(-16, 18, 1701)
def loss_mse_sigmoid(x, y=1):
  loss = 1/(1+np.exp(-x))
  loss = 1*(loss-1)**2
  return loss
loss = loss_mse_sigmoid(f,1)
plt.plot(f, loss)
plt.title(' loss: mse+sigmoid')
plt.grid(b=True)
````

## adagrad

````
def adagrad(X_train, Y_train, X_valid, Y_valid, W, alpha, num_iters):
    m = len(Y_train)
    train_loss = np.zeros((num_iters, 1))
    valid_loss = np.zeros((num_iters, 1))
    alpha_sqrt = np.zeros((4,1))
    alpha_ada = np.zeros((4,1))
 
    for i in range(num_iters):
        delta = gradient(function, W) 
        # output of function gradient is a list
        delta = np.array(delta)
        alpha_sqrt += delta**2
        alpha_ada += 1/np.sqrt(alpha_sqrt)
        W -= alpha*alpha_ada*delta.reshape(len(delta), 1)
 
        train_loss[i] = MSE(Y_train, np.matmul(X_train, W))
        valid_loss[i] = MSE(Y_valid, np.matmul(X_valid, W))
        if i%10 == 0:
            print('The training loss of the', i+1, 'epoch is', train_loss[i][0].round(4), 
                  ', and the validation loss of the', i+1, 'epoch is', valid_loss[i][0].round(4))
    return W, train_loss, valid_loss
````

## IOU

````
def bb_IOU(boxA, boxB):
    # 
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)

    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)

    # return the intersection over union value
    return iou
````


## Gini Impurity
 ### 定義 $gini = 1- \sum\limits_{i=i}^{k}\space p_i^{\space2}$


````
import numpy as np
def gini(ls = [3,4]):
  gini = 0
  for i in ls:
    gini += (i/sum(ls))**2 
  gini = 1-gini
  return float(f'{gini:.3f}')

x = gini([3,4])
print(f'this is a x output, {x}')
gini([3,3]) #0.5
gini([6,0]) #0
gini([4,2]) #0.444
gini([5,1]) #0.278
gini([1,5]) #0.278
````

## Entropy

$entropy = - \sum\limits_{i=i}^{k}\space p_i{\space} log\space p_i$


````
import numpy as np

def entropy(ls=[3,4]):
  en = 0
  for i in ls:
    en += (-1)*(i/sum(ls))*np.log2((i/sum(ls))+ 0.000000001)
  print(f'entropy of {ls} is, {en:.3f}' )
  return float(f'{en:.3f}')
````

About $log$

*  $2^2 = 4$
*  $2^3 = 8$
*  $2^4 = 16$
*  $2^5 = 32$

反問，
* 那麼 4 是 2 的幾次方, 記作 $log_2\space 4 = 2$
* 那麼 8 是 2 的幾次方, 記作 $log_2\space 8 = 3$
* 那麼 16 是 2 的幾次方, 記作 $log_2\space 16 = 4$
* 那麼 32 是 2 的幾次方, 記作 $log_2\space 32 = 5$

* $2^0 = 1$
* $2^{-1} = \frac{1}{2} \space \Rightarrow log_2\space\frac{1}{2} = -1 \space \Rightarrow \space - log_2\space 2 = -1$
* $2^{-2} = \frac{1}{4}\space \Rightarrow log_2\space\frac{1}{4} = -2 \space \Rightarrow \space - log_2\space 4 = -2$
* $2^{-1} \times 2^1 = 2^0 = 1$
* $log_2 \space \frac{1}{2} = - log_2 \space 2$
* $log_2 \space \frac{1}{p} = - log_2 \space p$

````
import matplotlib.pyplot as plt
import numpy as np

def relu(x = 0):
  x[x<=0] = 0
  return x
````

### $f(x): y = x + 1$


$x\_inception \space is \space y \space = \space 0 \space  \rightarrow \space  x = -1$

````
x = np.linspace(-10, 10, 201)
y = x + 1
axis_x = np.linspace(-10, 10, 201)
axis_x_value = 0* axis_x#np.dot(0, axis_x)
axis_y = np.linspace(0, 0, 201)
axis_y_value = np.linspace(-10, 10, 201)

plt.plot(x, y)
plt.plot(axis_x, axis_x_value, c='grey', lw=1)
plt.plot(axis_y, axis_y_value, c='grey', lw=1)
plt.text(3, 8, 'y = x + 1')
print(f"y_intercept is {y[100]}")
print(f"x_intercept is -1")
````

### $f(x): y = relu(x + 1)$


$x\_inception \space is \space y \space = \space 0 \space  \rightarrow \space  x = -1$

````
x = np.linspace(-10, 10, 201)
y = x + 1
y = relu(y)
axis_x = np.linspace(-10, 10, 201)
axis_x_value = 0* axis_x#np.dot(0, axis_x)
axis_y = np.linspace(0, 0, 201)
axis_y_value = np.linspace(-10, 10, 201)

plt.plot(x, y)
plt.plot(axis_x, axis_x_value, c='grey', lw=1)
plt.plot(axis_y, axis_y_value, c='grey', lw=1)
plt.text(3, 8, 'y = x + 1')
print(f"y_intercept is {y[100]}")
print(f"x_intercept is -1")
````

### $f(x): y = ax + b$


$x\_inception \space is \space y \space = \space 0 \space  \rightarrow \space  x = -b/a$

````
x1 = np.linspace(-10, 10, 201)
y1 = x1 + 8 
# x_inception is -8, and slope is 1
y1 = relu(y1)
x2 = np.linspace(-10, 10, 201)
y2 = 3*x2 + 3
# x_inception is -1, and slope is 3
y2 = relu(y2)
axis_x = np.linspace(-10, 10, 201)
axis_x_value = 0* axis_x#np.dot(0, axis_x)
axis_y = np.linspace(0, 0, 201)
axis_y_value = np.linspace(-10, 10, 201)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(axis_x, axis_x_value, c='grey', lw=1)
plt.plot(axis_y, axis_y_value, c='grey', lw=1)
plt.text(3, 8, 'y = x + 1')
print(f"y_intercept is {y[100]}")
print(f"x_intercept is -1")
````

````
plt.plot(x1, y1+y2)
plt.plot(axis_x, axis_x_value, c='grey', lw=1)
plt.plot(axis_y, axis_y_value, c='grey', lw=1)
plt.text(3, 8, 'y = x + 1')
print(f"y_intercept is {y[100]}")
print(f"x_intercept is -1")
````

````
# (40, 40)
import math
D0 = -0.5 * math.log(0.5, 2) - 0.5*math.log(0.5, 2)
D0
````

````
# (30, 10)
D1AL = (40/80)*(-0.75*math.log(0.75, 2) - 0.25*math.log(0.25,2))
# (10, 30)
D1AR = (40/80)*(-0.25*math.log(0.25, 2) - 0.75*math.log(0.75,2))

D1A = D1AL + D1AR
Entropy_A = D0 - D1A
print(D1AL, D1AR, D1A)
print(f'Entropy gain at case A: {Entropy_A}')
````

````
D1BL = (60/80)*(-(20/60)*math.log((20/60), 2) - (40/60)*math.log((40/60),2))
D1BR = (20/80)*(-math.log(1, 2))# - 0*math.log(0,2))
D1B = D1BL + D1BR
Entropy_B = D0 - D1B
print(D1BL, D1BR, D1B)
print(f'Entropy gain at case A: {Entropy_B}')
````
