```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x = tf.linspace(-10.,10,100)
y = tf.nn.sigmoid(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid()
plt.show()
```

```python
y = tf.nn.tanh(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("tanh Function")
plt.grid()
plt.show()
```

```python
y = tf.nn.relu(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Relu(x)")
plt.title("Relu Function")
plt.grid()
plt.show()
```

```python
y = tf.nn.leaky_relu(x, alpha = 0.2)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("leaky_relu(x)")
plt.title("leaky_relu Function")
plt.grid()
plt.show()
```

```python
y = tf.nn.elu(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.title("ELU Function")
plt.grid()
plt.show()
```
