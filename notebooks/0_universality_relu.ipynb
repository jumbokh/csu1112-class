```
import numpy as np
import matplotlib.pyplot as plt

def relu(x,b):
  y=np.zeros(x.shape[0])
  for i in range(x.shape[0]):
    if x[i]+b >0:
      y[i] = x[i]+b
    else:
      y[i] = 0
  return y

x1 = np.linspace(-100,100,201)
y1 = -1*relu(x1,100)+1*relu(x1,60)-3*relu(x1,29)-2*relu(x1,22)+3*relu(x1,0)+2*relu(x1,-20)+3*relu(x1,-50)
# 這個就是一個單一輸入 x 經過 7 個神經元 (權重 (1,1,1,1,1,1,1) bias(100, 60, 29, 22, 0, -20, -50)
# 然後再接一個神經元（權重（-1, 1, -3, -2, 3, 2, 3) 沒有 bias 得出 y

plt.plot(x1,y1)
```

```
import numpy as np

W = np.array([0.5, 1, 0.5, 0.5,1, -1, -1, 1.25, 1.25, -1.25, -1.25 ]).transpose()
B = [50, 40, 30, 20, 10, 0, -10, -20, -30, -40, -50]
data = np.linspace(-60, 60, 241 )

def universality(W, B, X):
  Z = np.dot(W, X) + B
  Z[Z<=0] = 0
  Z = np.sum(Z)
  return Z

universality(W, B, data[0])

out = []
for i in data:
  out.append(universality(W, B, i))

out

import matplotlib.pyplot as plt
plt.plot(data, out)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def relu(x,b):
  y=np.zeros(x.shape[0])
  for i in range(x.shape[0]):
    if x[i]+b >0:
      y[i] = x[i]+b
    else:
      y[i] = 0
  return y

x = np.linspace(-100,100,201)
y1 = 1*relu(x,100)
plt.plot(x,y1)
plt.xlabel('relu +1 at x= -100')
plt.ylim(-400,400)
plt.show()

y2 = -1*relu(x,80)
plt.plot(x,y2)
plt.xlabel('relu -1 at x= -80')
plt.ylim(-400,400)
plt.show()

y3 = y1+y2
plt.plot(x, y3)
plt.xlabel('relu -1 at x = -100, and +1 at x = -80')
plt.ylim(-400,400)
plt.show()

y4 = 1.5*relu(x,30) + y3
plt.plot(x,y4)
plt.xlabel('relu +1.5 at x= -30')
plt.ylim(-400,400)
plt.show()
```
