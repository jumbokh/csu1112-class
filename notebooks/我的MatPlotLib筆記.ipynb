# MatPlotLib 簡介

> MatPlotLib 是 <font color="red">Mat</font>rix <font color="red">Plot</font>ting <font color="red">Lib</font>ray 的縮寫。它是 Python 主要的 2D 繪圖套件之一，是 Python 繪圖的基礎套件。

> 在MatPlotLib中，我們通常用兩種方式來畫圖。第一種是用pyplot裡面的函式來做簡單的繪圖(模仿工程軟體MatLab的繪圖方式)，比較沒有彈性；另一種是使用Artist物件，用物件導向式的API來做更精確的繪圖。

## Markers, Line Styles, Colors

* [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

````python
x = np.arange(10, 20)
plt.plot(x,'k^--')
plt.plot(x**2,'ro-')
````

## LineWidth (lw), MarkerSize (ms), MarkerEdgeColor(mec), MarkerFaceColor (mfc)

````python
x = np.arange(10, 20)
plt.plot(x,'k^-', lw=3, ms=12, mec='k', mfc='w')
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow')
````

## Title, Xlabel, Ylabel

````python
x = np.arange(10, 20)
plt.plot(x,'k^-', lw=3, ms=12, mec='k', mfc='w')
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow')
plt.title('$x$ and $x^2$')
````

````python
x = np.arange(10, 20)
plt.plot(x,'k^-', lw=3, ms=12, mec='k', mfc='w')
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow')
plt.title('$x$ and $x^2$');
````

````python
x = np.arange(10, 20)
plt.plot(x,'k^-', lw=3, ms=12, mec='k', mfc='w')
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow')
plt.title('$x$ and $x^2$');
plt.xlabel('x')
plt.ylabel('y')
````

## Label & Legend

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

````python
x = np.arange(10, 20)
plt.plot(x,'k^-', lw=3, ms=12, mec='k', mfc='w', label='$x$')
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow', label='$x^2$')
plt.title('$x$ and $x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
````

## 子圖 (Subplots)

````python
x = np.arange(10)
plt.subplot(121)
plt.plot(x,'ks-', lw=3, ms=12, mec='k', mfc='w', label='$x$')
plt.subplot(122)
plt.plot(x**2,'ro-', lw=3, ms=12, mec='r', mfc='yellow', label='$x^2$')
````

````python
x = np.arange(10)
plt.subplot(221)
plt.plot(x,'ks-', label='$x$')
plt.subplot(222)
plt.plot(x**2,'ro-', label='$x^2$')
plt.subplot(223)
plt.plot(x**3,'bv-', label='$x^3$')
plt.subplot(224)
plt.plot(x**4,'gd-', label='$x^4$')
````

## 綜合練習

請畫出(0, 2$\pi$)間的$\sin$跟$\cos$函式。

並在 legend 上標出哪一條線是$\sin$、哪一條是$\cos$。

````python
x = np.linspace(0, 2*np.pi, 100)
````

````python
y_s = np.sin(x)
y_c = np.cos(x)
````

````python
plt.plot(x, y_s, 'r-', label='$sin(x)$')
plt.plot(x, y_c, 'b--', label='$cos(x)$')
plt.legend()
plt.title('Trigonometric Functions')
````
